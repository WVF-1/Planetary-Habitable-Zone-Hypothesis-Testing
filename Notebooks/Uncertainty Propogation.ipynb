```python
# Import Necessary Libraries and the Data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import shuffle

# Load the Data
df = pd.read_csv("exoplanets_science_ready.csv")
```

```python
# Monte Carlo Sim

n_sim = 1000

upper = df["planet_radius_upper_unc_earth_radius"]
lower = df["planet_radius_lower_unc_earth_radius"]
sigma = (upper.abs() + lower.abs()) / 2

# Optional: remove zero-uncertainty rows
mask = sigma > 0
radii = df.loc[mask, "planet_radius_earth_radius"]
sigma = sigma.loc[mask]

simulated_means = []

for _ in range(n_sim):
    sampled_radii = np.random.normal(
        loc=radii,
        scale=sigma
    )
    simulated_means.append(sampled_radii.mean())

np.percentile(simulated_means, [2.5, 97.5])
```

```python
# Permutation Testing


# Create composition label
df["composition"] = np.where(
    df["planet_radius_earth_radius"] <= 1.6,
    "Rocky",
    "Gaseous"
)

observed_diff = (
    df[df["composition"] == "Rocky"]["planet_radius_earth_radius"].mean()
    - df[df["composition"] == "Gaseous"]["planet_radius_earth_radius"].mean()
)

perm_diffs = []

for _ in range(n_sim):
    shuffled = shuffle(df["composition"])
    diff = (
        df[shuffled == "Rocky"]["planet_radius_earth_radius"].mean()
        - df[shuffled == "Gaseous"]["planet_radius_earth_radius"].mean()
    )
    perm_diffs.append(diff)

np.mean(np.abs(perm_diffs) >= abs(observed_diff))
```
