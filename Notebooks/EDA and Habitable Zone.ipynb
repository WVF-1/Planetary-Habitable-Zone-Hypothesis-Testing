```python
# Correlation Analysis
# Create a copy of the dataframe for correlation analysis
df_corr = df.copy()

# Enumerate discovery_method (convert categorical to numeric)
# This creates a mapping of unique methods to numbers
df_corr['discovery_method_numeric'] = (pd.factorize(df['discovery_method'])[0] + 1) # Starts at One

# Select only numeric columns for correlation
numeric_cols = df_corr.select_dtypes(include=[np.number]).columns
df_corr_numeric = df_corr[numeric_cols]

# Calculate correlations
correlation_matrix = df_corr_numeric.corr()

# Plot correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, annot_kws={'size': 9})
plt.title('Parameter Correlation Matrix', fontsize=14, pad=15)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()

# View the mapping of discovery methods to numbers
print("\nDiscovery Method Mapping:")
method_mapping = dict(enumerate(df['discovery_method'].unique()))
for num, method in method_mapping.items():
    print(f"{num}: {method}")
```

```python
# Missing Value Analysis

# Count missing values per column
na_counts = df.isna().sum()

# Convert to DataFrame for plotting
na_df = na_counts.reset_index()
na_df.columns = ['Variable', 'NA_Count']

# Plot histogram (bar chart) of NA counts
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(x='Variable', y='NA_Count', data=na_df, palette='viridis')

# Add labels and title
plt.title('Count of Missing Values per Variable', fontsize=14)
plt.xlabel('Variable', fontsize=12)
plt.ylabel('Number of Missing Values', fontsize=12)

# Show the plot
plt.show()
```

```python
# Basic overview
print("Shape:", df.shape)
df.info()

# Missingness summary
missing = df.isna().mean().sort_values(ascending=False)
print("\nMissingness (fraction):")
print(missing)


plt.style.use("dark_background")

# -----------------------------
# Univariate distributions
# -----------------------------

numeric_cols = [
    "planet_radius_earth_radius",
    "planet_mass_earth_mass",
    "orbital_period_days",
    "insolation_flux_earth_flux",
    "equilibrium_temperature_k"
]

for col in numeric_cols:
    plt.figure()
    df[col].dropna().plot(kind="hist", bins=40, log=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count (log scale)")
    plt.show()

# -----------------------------
# Bivariate relationships
# -----------------------------

plt.figure()
plt.scatter(df["planet_mass_earth_mass"], df["planet_radius_earth_radius"], alpha=0.3)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Planet Mass (Earth Masses)")
plt.ylabel("Planet Radius (Earth Radii)")
plt.title("Planet Mass vs Radius")
plt.show()

plt.figure()
plt.scatter(df["orbital_period_days"], df["insolation_flux_earth_flux"], alpha=0.3)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Orbital Period (days)")
plt.ylabel("Insolation (Earth Flux)")
plt.title("Orbital Period vs Insolation")
plt.show()

plt.figure()
plt.scatter(df["insolation_flux_earth_flux"], df["equilibrium_temperature_k"], alpha=0.3)
plt.xscale("log")
plt.xlabel("Insolation (Earth Flux)")
plt.ylabel("Equilibrium Temperature (K)")
plt.title("Temperature vs Insolation")
plt.show()

# -----------------------------
# Detection method bias
# -----------------------------

method_counts = df["discovery_method"].value_counts()
method_counts.plot(kind="bar")
plt.title("Planet Counts by Discovery Method")
plt.ylabel("Count")
plt.show()
```

```python
# -----------------------------
# Habitable Zone Definition
# -----------------------------
# Using conservative insolation-based definition
# (Earth ~1.0; typical conservative bounds ~0.251.75)

HZ_LOWER = 0.25
HZ_UPPER = 1.75

df["in_habitable_zone"] = (
    (df["insolation_flux_earth_flux"] >= HZ_LOWER) &
    (df["insolation_flux_earth_flux"] <= HZ_UPPER)
)

print(df["in_habitable_zone"].value_counts())


plt.style.use("dark_background")

# -----------------------------
# Composition Classification
# -----------------------------
# Radius-based classification
# Rocky < 1.6 Earth radii

ROCKY_THRESHOLD = 1.6

df["composition"] = np.where(
    df["planet_radius_earth_radius"] < ROCKY_THRESHOLD,
    "Rocky",
    "Gaseous"
)

print(df["composition"].value_counts())

# -----------------------------
# Stratified Groups
# -----------------------------

strata = df.groupby(["composition", "in_habitable_zone"]).size()
print("\nCounts by Composition and HZ:")
print(strata)

# -----------------------------
# Visualization
# -----------------------------

plt.figure()
for comp in ["Rocky", "Gaseous"]:
    subset = df[df["composition"] == comp]
    plt.scatter(
        subset["insolation_flux_earth_flux"],
        subset["planet_radius_earth_radius"],
        alpha=0.4,
        label=comp
    )

plt.axvline(HZ_LOWER, linestyle="--")
plt.axvline(HZ_UPPER, linestyle="--")
plt.xscale("log")
plt.xlabel("Insolation (Earth Flux)")
plt.ylabel("Planet Radius (Earth Radii)")
plt.title("Planet Radius vs Insolation with Habitable Zone")
plt.legend()
plt.show()
```
